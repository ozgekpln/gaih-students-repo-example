```
import random
import string
kelimeler=["istanbul","erzurum","trabzon","siirt","izmir","adana","bilgisayar","sivas","fransa","italya","türkiye","klavye","televizyon","sinema","tiyatro","üniversite"]
def alinan_kelime(kelimeler):
  kelime=random.choice(kelimeler)
  while "-" in kelime or " " in kelime:
    kelime=random.choice(kelimeler)
  return kelime.upper()
#def kelime_harfleri():
 # kelime_harfleri=set(kelime)
      
class hangman:
  def __init__(self):
    self.kelime=alinan_kelime(kelimeler)
    self.kelimeHarfleri= set(self.kelime)
    self.alfabe=set(string.ascii_uppercase)
    self.tahminiHarfler=set()
    self.can=6
        
        
    while len(self.kelimeHarfleri)>0 and self.can>0:
      print(self.can," canın kaldı. ","Şimdiye kadar şu harfleri tahmin ettin: ","".join(self.tahminiHarfler))
      word_list=[harf if harf in self.tahminiHarfler else "-" for harf in self.kelime]
      print("şu anlık kelime: ","".join(word_list))
            
            
      harfTahmini=input("Bir harf giriniz: ").upper()
      if harfTahmini in self.alfabe-self.tahminiHarfler:
        self.tahminiHarfler.add(harfTahmini)
        if harfTahmini in self.kelimeHarfleri:
          self.kelimeHarfleri.remove(harfTahmini)
          print("")
        else:
          self.can=self.can-1
          print("\nTahmin ettiğin harf kelimenin içinde yok.")
      elif harfTahmini in self.tahminiHarfler:
        print("Bu harfi daha önce tahmin ettin. Başka bir harf tahmin et.")
      else:
        print("Hatalı bir tahminde bulundunuz.")     
    if self.can==0:
      print("ADAM ASILDI!!! Tahmin etmen gereken kelime: ",self.kelime)
    else:
      print("TEBRİKLERRR KELİMEYİ BİLDİNİZ:)",self.kelime) 

if __name__=='__main__':
  oyun=hangman()

  
        
    	
```
